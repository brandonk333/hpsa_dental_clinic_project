```python
import os
import pandas as pd
```

```python
data_file_folder = '.\hpsa_data_export'

df = []
for file in os.listdir(data_file_folder):
    if file.endswith('.xlsx'):
        print('Loading file {0}...'.format(file))
        df.append(pd.read_excel(os.path.join(data_file_folder, file), sheet_name = 'HPSA Find'))
```

```python
len(df)
```

```python
df_master = pd.concat(df, axis=0)
```

```python
df_master
```

```python
df_master = df_master.rename(columns = {'Unnamed: 0':'Discipline','Unnamed: 1':'HPSA_ID','Unnamed: 2':'HPSA_Name','Unnamed: 3':'Designation_Type',
                           'Unnamed: 4':'State','Unnamed: 5':'County','Unnamed: 6':'HPSA_FTE_Short','Unnamed: 7':'HPSA_Score',
                           'Unnamed: 8':'PC_MCTA_Score','Unnamed: 9':'Status','Unnamed: 10':'Rural_Status','Unnamed: 11':'Designation_Date',
                           'Unnamed: 12':'Update_Date'})
```

```python
df_master
```

```python
df_master = df_master.drop([0,1,2],axis=0)
```

```python
df_master
```

```python
df_master = df_master[df_master.HPSA_ID.str.startswith('6', na=False)]
```

```python
df_master
```

```python
df_master = df_master.drop_duplicates(subset=['HPSA_ID'])
```

```python
df_master.head(2)
```

```python
isinstance(df_master,pd.DataFrame)
```

```python
df_master = df_master.reset_index()
```

```python
df_master.drop('index', axis=1, inplace=True)
```

```python
df_master.head(2)
```

```python
df_master.loc[825:835]
```

```python
df_master.dropna(subset=['Discipline'], inplace=True)
```

```python
for column in df_master.columns:
    df_master[column] = df_master[column].str.strip()
    df_master[column] = df_master[column].str.title()
```

```python
df_master.drop(['Discipline','HPSA_FTE_Short','PC_MCTA_Score','Designation_Date'], axis=1, inplace=True)
```
